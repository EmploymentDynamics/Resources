## Tech RF and LR Model

```python
# make balanced data
#Split data by age group
from sklearn.linear_model import LogisticRegression

agegroups=[1,2,3,4]
agegroupstring=['1:> 45 yrs old','2:36-45 yrs old','3:26-35 yrs old','4:< 26 yrs old']
rf_dataframe_age1=pd.DataFrame()
rf_dataframe_age2=pd.DataFrame()
rf_dataframe_age3=pd.DataFrame()
rf_dataframe_age4=pd.DataFrame()

lr_dataframe_age1=pd.DataFrame()
lr_dataframe_age2=pd.DataFrame()
lr_dataframe_age3=pd.DataFrame()
lr_dataframe_age4=pd.DataFrame()

for x in range(len(agegroups)):
    temp=dataTech
    temp=temp[dataTech['Age']==agegroups[x]]
    not_laid_off_tech = temp[temp['laid_off'] == 0]
    laid_off_tech = temp[temp['laid_off'] == 1]
    np.random.seed(101)
    not_laid_off_tech = not_laid_off_tech.sample(n=len(laid_off_tech)) 
    temp = pd.concat([not_laid_off_tech, laid_off_tech])

    from sklearn.model_selection import train_test_split
    # Reserve 20% for testing
    y = temp['laid_off']
    X = temp.drop(['Length', 'ID','Birth Year','City of profile', 'Country of profile',
                       'Start Date','End Date','Role', 'Dept', 'Company Name',
                       'Location','Industry','laid_off','Layoff Date', 'Employees Affected'],axis = 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=101)

    print('training data has %d observation with %d features'% X_train.shape)
    print('test data has %d observation with %d features'% X_test.shape)

    # Scale the data, using standardization
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Random forest to see feature importance
    from sklearn.ensemble import RandomForestClassifier
    RF = RandomForestClassifier()
    RF.fit(X_train, y_train)
    RF.score(X_test,y_test)

    # hypermarameter tuning for Random Forest
    from sklearn.model_selection import GridSearchCV
    parameters = {
        'n_estimators' : [40,60,80]
    }
    Grid_RF = GridSearchCV(RandomForestClassifier(),parameters, cv=5)
    Grid_RF.fit(X_train, y_train)
    
    # feature importance
    importances = RF.feature_importances_
    # Print the feature ranking
    #print("Feature importance ranking by Random Forest Model:")
   # for k,v in sorted(zip(map(lambda x: round(x, 4), importances), X.columns), reverse=True):
    #    print (v + ": " + str(k))
    
    LR = LogisticRegression()
    LR.fit(X_train, y_train)
    LR.score(X_test, y_test)

    parameters = {
        'penalty':('l1', 'l2'), 
        'C':(1, 5, 10)
    }
    Grid_LR = GridSearchCV(LogisticRegression(),parameters, cv=5)
    Grid_LR.fit(X_train, y_train)

    best_LR_model = Grid_LR.best_estimator_

    scaler = StandardScaler()
    X_l1 = scaler.fit_transform(X)
    best_LR_model.fit(X_l1, y)
    best_LR_model.coef_[0]
    print ("Logistic Regression (L1) Coefficients")

    if x==0:
        rf_dataframe_age1['Features']=X.columns
        rf_dataframe_age1['Importance Score:1:> 45 yrs old']=RF.feature_importances_
        rf_dataframe_age1=rf_dataframe_age1.sort_values(by='Importance Score:1:> 45 yrs old', ascending=False)
        lr_dataframe_age1['Features']=X.columns
        lr_dataframe_age1['LR Coefficients:1:> 45 yrs old']=best_LR_model.coef_[0]
        lr_dataframe_age1=lr_dataframe_age1.sort_values(by='LR Coefficients:1:> 45 yrs old', ascending=False)

    
    if x==1:
        rf_dataframe_age2['Features']=X.columns
        rf_dataframe_age2['Importance Score:2:36-45 yrs old']=RF.feature_importances_
        rf_dataframe_age2=rf_dataframe_age2.sort_values(by='Importance Score:2:36-45 yrs old', ascending=False)
        lr_dataframe_age2['Features']=X.columns
        lr_dataframe_age2['LR Coefficients:2:36-45 yrs old']=best_LR_model.coef_[0]
        lr_dataframe_age2=lr_dataframe_age2.sort_values(by='LR Coefficients:2:36-45 yrs old', ascending=False)

    
    if x==2:
        rf_dataframe_age3['Features']=X.columns
        rf_dataframe_age3['Importance Score:3:26-35 yrs old']=RF.feature_importances_
        rf_dataframe_age3= rf_dataframe_age3.sort_values(by='Importance Score:3:26-35 yrs old', ascending=False)
        lr_dataframe_age3['Features']=X.columns
        lr_dataframe_age3['LR Coefficients:3:26-35 yrs old']=best_LR_model.coef_[0]
        lr_dataframe_age3=lr_dataframe_age3.sort_values(by='LR Coefficients:3:26-35 yrs old', ascending=False)

    
    if x==3:
        rf_dataframe_age4['Features']=X.columns
        rf_dataframe_age4['Importance Score:4:< 26 yrs old']=RF.feature_importances_
        rf_dataframe_age4=rf_dataframe_age4.sort_values(by='Importance Score:4:< 26 yrs old', ascending=False)
        lr_dataframe_age4['Features']=X.columns
        lr_dataframe_age4['LR Coefficients:4:< 26 yrs old']=best_LR_model.coef_[0]
        lr_dataframe_age4=lr_dataframe_age4.sort_values(by='LR Coefficients:4:< 26 yrs old', ascending=False)

    

    
```
