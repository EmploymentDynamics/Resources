```python
import string
```

```python
# Group the number of laid off employees acorss different branches under different layoff events from the same company
summary_wo_date = summary_w_date.groupby(['Company Name']).sum().reset_index()
summary_wo_date.head(10)
```

```python
#Initial grouping. More manuel work needed later for grouping the same company  with different name formats

#Uniformize the upper case letters
summary_wo_date['Company Name'] = [i.upper() for i in summary_wo_date['Company Name']]

# Remove Punctuation
translator = str.maketrans('', '', string.punctuation)

for i in range(summary_wo_date.shape[0]):    
    summary_wo_date['Company Name'][i] = summary_wo_date['Company Name'][i].translate(translator)
    

summary_wo_date = summary_wo_date.groupby(['Company Name']).sum().sort_values(by=['Employees Affected'], ascending = False)
summary_wo_date.head(10)
```
